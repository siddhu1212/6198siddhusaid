**UNIVARIATE ANALYSIS:**

```python
#1. histogram
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='Age Range', bins=10, kde=True)
plt.title('Histogram of Age Range')
plt.xlabel('Age Range')
plt.ylabel('Frequency')
plt.show()
```

insights:


    Data Distribution: The histogram provides an overview of the distribution of ages among the patrons in your dataset.

    Age Range Groups: The x-axis represents the different age range groups, while the y-axis represents the frequency or count of patrons in each age range.

    Shape of the Distribution: The histogram shows that the distribution of ages appears to be somewhat right-skewed, with the majority of patrons falling into the younger age ranges.

    Peak Age Range: The age range with the highest frequency appears to be in the younger age group, suggesting that a significant portion of the patrons are relatively young.

    Outliers: You can also observe whether there are any outliers or unusual patterns in the data, which may be represented as tall bars far from the main cluster of data.

```python
#2. box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, y='Total Checkouts')
plt.title('Box Plot of Total Checkouts')
plt.ylabel('Total Checkouts')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
data['Home Library Definition'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Home Library Definition Distribution')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Circulation Active Year', order=data['Circulation Active Year'].value_counts().index)
plt.title('Circulation Active Year Count')
plt.xlabel('Circulation Active Year')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

# **BIVARIATE ANALYSIS:**




```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='Total Checkouts', y='Total Renewals')
plt.title('Scatter Plot of Total Checkouts vs. Total Renewals')
plt.xlabel('Total Checkouts')
plt.ylabel('Total Renewals')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='Home Library Definition', y='Total Checkouts')
plt.title('Box Plot of Total Checkouts by Home Library Definition')
plt.xlabel('Home Library Definition')
plt.ylabel('Total Checkouts')
plt.xticks(rotation=90)
plt.show()
```

```python
yearly_registration_counts = data['Year Patron Registered'].value_counts().sort_index()

# Create a line chart for yearly patron registration counts
plt.figure(figsize=(10, 6))
plt.plot(yearly_registration_counts.index, yearly_registration_counts.values, marker='o', linestyle='-')
plt.title('Yearly Patron Registration Trends')
plt.xlabel('Year Patron Registered')
plt.ylabel('Number of Patrons')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**MULTIVARIATE ANALYSIS:**

```python
plt.figure(figsize=(10, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(data=data[['Total Checkouts', 'Total Renewals', 'Age Range']], diag_kind='kde')
plt.suptitle('Pairplot for Numeric Variables')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='Patron Type Definition', y='Total Checkouts', order=data['Patron Type Definition'].value_counts().index)
plt.title('Box Plot of Total Checkouts by Patron Type Definition')
plt.xlabel('Patron Type Definition')
plt.ylabel('Total Checkouts')
plt.xticks(rotation=90)
plt.show()
```
